```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

plt.style.use('seaborn')
pd.set_option('display.float_format',lambda x : '%.2f' % x)
```

```python
train = json.load(open('../data/train.json'))
test = json.load(open('../data/test.json'))
train = pd.DataFrame(train).T
test = pd.DataFrame(test).T
```

```python
train['author_num'] = [len(author) for author in train['author']]
count = train.groupby('author_num').size()
total = [c for c in count]
number = [i for i, c in zip(count.index, count)]
```

```python
fig = plt.figure(figsize = (3, 3))
plt.bar(number, total, width=0.4)
plt.xlabel("Number of Co-authors")
plt.ylabel("Count")
plt.title("Frequency Chart of Number of Co-author")
plt.show()
```

```python
all_authors = []
for author in train['author']:
    all_authors.extend(author)
all_authors = Counter(all_authors)
author_freq = sorted(list(all_authors.values()), reverse=True)
```

```python
fig = plt.figure(figsize = (3, 3))
plt.plot(author_freq)
plt.ylabel('Number of papers written')
plt.title('Co-author Frequency for Each Author')
plt.plot()
```

```python
all_keywords = []
for keyword in train['keywords']:
    all_keywords.extend(keyword)
all_keywords = Counter(all_keywords)
keyword_freq = sorted(list(all_keywords.values()), reverse=True)
```

```python
fig = plt.figure(figsize = (3, 3))
plt.plot(keyword_freq)
plt.title('Keywords Frequency for All Paper')
plt.ylabel('Frequency Count')
plt.show()
```
