```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

plt.style.use('seaborn')
pd.set_option('display.float_format',lambda x : '%.2f' % x)
```

```python
train = json.load(open('../data/train.json'))
test = json.load(open('../data/test.json'))
train = pd.DataFrame(train).T
test = pd.DataFrame(test).T
```

```python
train['author_num'] = [len(author) for author in train['author']]
count = train.groupby('author_num').size()
total = [c for c in count]
number = [i for i, c in zip(count.index, count)]
```

```python
fig = plt.figure(figsize = (4, 3))
plt.bar(number, total, width=0.4)
plt.xlabel("Number of Co-authors")
plt.ylabel("Number of Paper")
plt.title("Paper Frequency Chart of Number of Co-author")
plt.savefig('freq_count_aoauthor.png')
plt.show()
```

```python
keywords_count = train.keywords.str.len()
```

```python
fig = plt.figure(figsize = (4, 3))
sns.histplot(keywords_count)
plt.title('Keywords Count for All Paper')
plt.savefig('keywords_count.png')
plt.ylabel('Number of Papers')
plt.xlabel('Number of Keywords')
plt.savefig('keywords_count.png')
plt.show()
```

```python
author_count = train.author.str.len()
keywords_count = train.keywords.str.len()

fig = plt.figure(figsize = (5, 3))
sns.boxplot(x=author_count, y=keywords_count)
plt.title('Keyword Count  VS Author Count')
plt.ylabel('Number of Keyword')
plt.xlabel('Number of Author')
plt.savefig('author_vs_keyword.png')
plt.show()
```

```python
all_authors = []
for author in train['author']:
    all_authors.extend(author)
all_authors = Counter(all_authors)
author_freq = sorted(list(all_authors.values()), reverse=True)
```

```python
fig = plt.figure(figsize = (5, 3))
plt.plot(author_freq)
plt.ylabel('Number of papers written')
plt.title('Number of papers written per author (Sorted)')
plt.plot()
```
